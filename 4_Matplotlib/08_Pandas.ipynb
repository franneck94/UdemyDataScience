# Pandas Visualization 

```python
import matplotlib.pyplot as plt
import numpy as np


np.random.seed(0)
import pandas as pd
```

```python
num_samples = 100

ts = pd.Series(
    np.random.randn(num_samples),
    index=pd.date_range("1/1/2000", periods=num_samples),
)
```

```python
print(ts.head())
```

```python
ts.plot()
plt.show()
```

```python
num_samples = 100

data = np.concatenate(
    [
        np.random.randn(num_samples, 2),
        np.random.randint(low=0, high=10, size=(num_samples, 1)),
    ],
    axis=1,
)

df = pd.DataFrame(data, index=ts.index, columns=["A", "B", "C"]).astype(
    {"A": np.float32, "B": np.float32, "C": np.uint8}
)
```

```python
print(df.head())
```

```python
df.plot()
plt.show()
```

Plotting methods allow for a handful of plot styles other than the default line plot. These methods can be provided as the kind keyword argument to plot(), and include:

- ‘bar’ or ‘barh’ for bar plots
- ‘hist’ for histogram
- ‘box’ for boxplot
- ‘kde’ or ‘density’ for density plots
- ‘area’ for area plots
- ‘scatter’ for scatter plots
- ‘hexbin’ for hexagonal bin plots
- ‘pie’ for pie plots

```python
df.plot(kind="box", x="A", y="B")
plt.show()
```

```python
df.plot(kind="hist")
plt.show()
```

```python
df.plot(kind="kde")
plt.show()
```

```python
df.plot(kind="scatter", x="A", y="B", marker="*")
plt.show()
```

```python
df.plot(kind="hexbin", x="A", y="B")
plt.show()
```
